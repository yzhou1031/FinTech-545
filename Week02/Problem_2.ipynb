```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
input_path = "test7_2.csv"
output_path = "testout7_2.csv"
```

```python
df = pd.read_csv(input_path)
x = pd.to_numeric(df["x1"], errors="coerce").dropna().values
```

```python
nu, mu, sigma = stats.t.fit(x)

result_df = pd.DataFrame({"mu": [mu], "sigma": [sigma], "nu": [nu]})
result_df.to_csv(output_path, index=False)
```
