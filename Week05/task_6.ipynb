```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("test6.csv")
P = df.select_dtypes(include="number")
```

```python
arith_ret = P.pct_change()
arith_ret = arith_ret.dropna()
arith_ret.to_csv("testout_6.1.csv", index=False)
print("Task 6.1 - arith_ret:\n")
arith_ret
```

```python
log_ret = np.log(P / P.shift(1))
log_ret = log_ret.dropna()
log_ret.to_csv("testout_6.2.csv", index=False)
print("Task 6.2 - log_ret:\n")
log_ret
```
