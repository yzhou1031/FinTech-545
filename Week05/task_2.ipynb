```python
import pandas as pd
import numpy as np

df = pd.read_csv("test2.csv")
R = df.select_dtypes(include="number")
```

```python
lam_1 = 0.97
alpha_1 = 1 - lam_1

ew_cov_097 = (
    R
    .ewm(alpha=alpha_1, adjust=True)
    .cov(bias=True)
    .xs(R.index[-1])
)
print("Task 2.1 - ew_cov_097:\n", ew_cov_097)
```

```python
lam_2 = 0.94
alpha_2 = 1 - lam_2

ew_cov_094 = (
    R
    .ewm(alpha=alpha_2, adjust=True)
    .cov(bias=True)
    .xs(R.index[-1])
)

ew_var_094 = (
    R
    .ewm(alpha=alpha_2, adjust=True)
    .var(bias=True)
    .iloc[-1]
)

sigma_094 = np.sqrt(ew_var_094)

ew_corr_094 = ew_cov_094.mul(1 / sigma_094, axis=0).mul(1 / sigma_094, axis=1)

print("Task 2.2 - ew_corr_094:\n", ew_corr_094)
```

```python
lam_097 = 0.97
alpha_097 = 1 - lam_097

ew_var_097 = (
    R
    .ewm(alpha=alpha_097, adjust=True)
    .var(bias=True)
)

last_var = ew_var_097.iloc[-1]
sigma_097 = np.sqrt(last_var) 
cov_hybrid = ew_corr_094.mul(sigma_097, axis=0).mul(sigma_097, axis=1)

print("Task 2.3 - cov_hybrid:\n", cov_hybrid)
```
