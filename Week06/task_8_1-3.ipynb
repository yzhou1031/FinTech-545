```python
import numpy as np
import pandas as pd
from scipy.stats import norm
from scipy.stats import t
```

```python
df = pd.read_csv("test7_1.csv")
data = df.iloc[:, 0].dropna().values

mu = np.mean(data)
sigma = np.std(data, ddof=1)   # most assignments expect ddof=1

alpha = 0.95
z = norm.ppf(alpha)

VaR_from_mean = abs(z) * sigma
VaR_absolute  = abs(mu - z * sigma)

print("Task 8.1 - VaR from Normal:")
print(f"VaR Diff from Mean ({alpha*100:.0f}%):", VaR_from_mean)
print(f"VaR Absolute ({alpha*100:.0f}%):", VaR_absolute)
```

```python
df = pd.read_csv("test7_2.csv")
data = df.iloc[:, 0].dropna().values

alpha = 0.95
nu, mu, scale = t.fit(data)

t_alpha = t.ppf(alpha, df=nu)

VaR_from_mean_t = abs(t_alpha) * scale
VaR_absolute_t  = abs(mu - VaR_from_mean_t)

print("Task 8.2 - VaR from T:")
print(f"VaR Diff from Mean ({alpha*100:.0f}%):", VaR_from_mean_t)
print(f"VaR Absolute ({alpha*100:.0f}%):", VaR_absolute_t)
```

```python
np.random.seed(4)

n_sims = 100000

simulated = t.rvs(df=nu, loc=mu, scale=scale, size=n_sims)

q = np.quantile(simulated, 1 - alpha)

VaR_from_mean_sim = abs(q - np.mean(simulated))
VaR_absolute_sim  = abs(q)

print("Task 8.3 - VaR from Simulation:")
print(f"VaR Diff from Mean ({alpha*100:.0f}%):", VaR_from_mean_sim)
print(f"VaR Absolute ({alpha*100:.0f}%):", VaR_absolute_sim)
```
